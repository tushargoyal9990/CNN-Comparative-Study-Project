**Import Statements**

```
import matplotlib.pyplot as plt
import numpy as np
```

**Classes**

```
x = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

**Precision Plot**

```
y1 = [90, 86, 83, 78, 84, 81, 90, 93, 93, 91] # SequentialNet
y2 = [84, 88, 68, 66, 71, 75, 75, 84, 91, 86] # InceptionNet
y3 = [77, 90, 60, 78, 90, 73, 94, 89, 78, 84] # DenseNet
y4 = [86, 89, 81, 71, 85, 81, 81, 87, 92, 92] # WideResNet
plt.plot(x, y1)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.scatter(x, y2)
plt.plot(x, y3)
plt.scatter(x, y3)
plt.plot(x, y4)
plt.scatter(x, y4)
plt.title('Classwise Precision Plot')
plt.ylabel('Precision')
plt.xlabel('Class')
plt.legend(['SequentialNet', 'InceptionNet', 'DenseNet', 'WideResNet'], bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation = 60)
plt.ylim(bottom=0, top=100)
plt.show()
```

**Recall Plot**

```
y1 = [87, 94, 81, 73, 88, 79, 92, 90, 92, 92] # SequentialNet
y2 = [81, 89, 73, 57, 80, 65, 89, 84, 84, 85] # InceptionNet
y3 = [81, 93, 88, 52, 68, 74, 77, 83, 92, 91] # DenseNet
y4 = [86, 94, 76, 73, 81, 73, 92, 89, 91, 89] # WideResNet
plt.plot(x, y1)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.scatter(x, y2)
plt.plot(x, y3)
plt.scatter(x, y3)
plt.plot(x, y4)
plt.scatter(x, y4)
plt.title('Classwise Recall Plot')
plt.ylabel('Recall')
plt.xlabel('Class')
plt.legend(['SequentialNet', 'InceptionNet', 'DenseNet', 'WideResNet'], bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation = 60)
plt.ylim(bottom=0, top=100)
plt.show()
```

**F1-Score Plot**

```
y1 = [89, 90, 82, 75, 86, 80, 91, 91, 93, 91] # SequentialNet
y2 = [82, 88, 71, 61, 75, 70, 81, 84, 87, 86] # InceptionNet
y3 = [79, 91, 71, 62, 77, 73, 85, 86, 85, 88] # DenseNet
y4 = [86, 91, 79, 72, 83, 77, 86, 88, 91, 90] # WideResNet
plt.plot(x, y1)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.scatter(x, y2)
plt.plot(x, y3)
plt.scatter(x, y3)
plt.plot(x, y4)
plt.scatter(x, y4)
plt.title('Classwise F1-Score Plot')
plt.ylabel('F1-Score')
plt.xlabel('Class')
plt.legend(['SequentialNet', 'InceptionNet', 'DenseNet', 'WideResNet'], bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation = 60)
plt.ylim(bottom=0, top=100)
plt.show()
```
